```python
# we will transfer that text to html code
soup = BeautifulSoup(response.text, 'html.parser')
print(soup)
```

```python
# we will get the class contain the data that i want
data = soup.find_all('div', attrs={'class':'clp-lead__badge-ratings-enrollment'})
for d in data:
    print(d)
    print('\n')
```

```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import time

while True:
    url = 'https://www.udemy.com/course/python-for-data-science-and-machine-learning-bootcamp/'
    response = requests.request('GET', url)

    soup = BeautifulSoup(response.text, 'html.parser')

    data = soup.find_all('div', attrs={'class':'clp-lead__badge-ratings-enrollment'})

    data_a = data[0].find('div').text

    nu = data_a.split()

    students = nu[7]
    rating = nu[1]

    print(students)
    print(rating)
    
    f = open('rating.txt', 'a')
    now = datetime.now()
    time_now = now.strftime("%d/%m/%Y")
    output = 'Total students= '+students+','+'Rating= '+rating+', at '+time_now+'\n\n'
    f.write(output)
    f.close()
    time.sleep(86400)
```
